```python
import unittest

review_matrix_one_week_three = [5, 4, 3, 2], [9, 7, 6, 5], [3, 9, 12, 15], [22, 18, 16, 10]
review_matrix_two_week_three = [18, 56, 13, 12], [7, 6, 5, 4], [33, 91, 25, 44], [21, 19, 2, 1]
review_vector_week_three = [7], [11], [10], [23]
review_add_matrix_result = [[23, 60, 16, 14], [16, 13, 11, 9], [36, 100, 37, 59], [43, 37, 18, 11]]
review_subtract_matrix_result = [[-13, -52, -10, -10], [2, 1, 1, 1], [-30, -82, -13, -29], [1, -1, 14, 9]]
review_multiply_matrix_result = [[259, 615, 164, 210], [514, 1187, 312, 405], [828, 1599, 414, 615],
                                 [1260, 2986, 796, 1050]]
review_multiply_by_vector_result = [[155], [315], [585], [742]]

week_four_determinent_matrix = [9, 7], [3, 2]
week_four_cross_product_tensor_one = [3, -5, 0]
week_four_cross_product_tensor_two = [2, 1, 9]
week_four_matrix_one = [9, 7, 6, 5], [8, 3, 2, 1], [9, 11, 12, 13], [1, 3, 5, 7]
week_four_matrix_two = [3, 2, 1, 0], [9, 1, 9, 0], [7, 6, 5, 4], [3, 1, 12, 7]
week_four_vector = [3], [2], [9], [4]

"""A exception used for raising exceptions for class"""


class TensorException(Exception):

    def __init__(self, error_message):
        self.error_message = error_message

class Tensor(BaseTensor):
    
    tensor_one = []
    tensor_two = []

    def __init__(self, tensor_one=[], tensor_two=[]):
        self.tensor_one = tensor_one
        self.tensor_two = tensor_two
        pass
        

    def get_inverse_tensor(self, tensor):
        """Given a tensor (Matrix) gets the inverse of matrix
            Args:
            tensor (list)  The values for the first
            Returns: (list)   A list containing the results of the inverse operation.
            Raises:
            TensorException"""
        try:
            determinent_dictionary = self.tensor_to_variables(tensor)
            tensor_determinant = self.get_determinent(tensor)
            if self.is_two_by_two_tensor(tensor):
                return [[self.divide(determinent_dictionary['d'], tensor_determinant),
                         -1 * self.divide(determinent_dictionary['b'], tensor_determinant)],
                        [-1 * self.divide(determinent_dictionary['c'], tensor_determinant),
                         self.divide(determinent_dictionary['a'], tensor_determinant)]]
        except:
            raise TensorException(
                "The inverse matrix is unable to be calculated for the defined tensor of" + str(tensor))

    def get_determinent(self, tensor):
        """Given a tensor (Matrix) gets the inverse of matrix
           Args:
           tensor (list)  The values of a 4x1 tensor
           Returns:
           answer (variable) The determinent variable
           Raises:
           TensorException """
        try: 
            determinent_dictionary = self.tensor_to_variables(tensor)
            a = determinent_dictionary['a']
            b = determinent_dictionary['b']
            c = determinent_dictionary['c']
            d = determinent_dictionary['d']
            answer = (a * d) - (b * c)
            return answer
        except:
            raise TensorException(
                "The determinent is unable to be found")

    def is_tensor_equal(self, tensor):
        """Given a tensor checks the matrix of equal length
        Args:
       tensor (list)  The values of a 4x4 tensor
       Returns:
       bool (variable) The result of whether the tensor is of equal size
       Raises:
       TensorException"""

        size = len(tensor[0])
        for element in tensor:
            if size != len(element):
                return False
        return True

    def add_or_subtract_tensor(self, operator=1):
        """Given a tensor (Matrix) adds or subtracts tensor
           Args:
           operator (variable) minus one is subtract and one is add
            Returns:
       tensor (list) The result of add or subtract operation
       Raises:
       TensorException"""
        try:
            # go no further if Matrices don't match in size
            tensor_one_size = self.find_tensor_size(self.tensor_one)
            tensor_two_size = self.find_tensor_size(self.tensor_two)
            if tensor_one_size != tensor_two_size:
                raise TensorException("Tensors DO NOT MATCH. Cannot perform operation.")

            # Create a blank matrix for the answr set all cells to 0
            answer_tensor = self.create_blank_tensor(tensor_one_size[0], tensor_two_size[1])
            for i in range(tensor_one_size[0]):
                for j in range(tensor_one_size[1]):
                    answer_tensor[i][j] = self.tensor_one[i][j] + (self.tensor_two[i][j] * operator)
            return answer_tensor
        except:
            raise TensorException(
                "The tensor is unable to be subtracted/added")

    def get_cross_product(self, vector_one, vector_two):
        """ Calculates the cross product according to formula
            (aybz - azby) = i
            (azbx - axbz) = j
            (axby - aybx) = k
            where vector_one is a and vector_two is b
            Args:
            vector_one (list)  The values of a 4x1 tensor
            vector_two (list)  The values of a 4x1 tensot
            Returns:
            result (tuple) The results og calculations for i , j , k
            Raises:
            TensorException"""
        try:
            if self.is_three_by_one_tensor(vector_one) and self.is_three_by_one_tensor(vector_two):
                a = self.tensor_to_variables(vector_one)
                b = self.tensor_to_variables(vector_two)

                element_i = self.get_cross_product_for_i(self.to_int(a['y']), self.to_int(b['z']),
                                                self.to_int(a['z']), self.to_int(b['y']))

                element_j = self.get_cross_product_for_j(self.to_int(a['z']), self.to_int(b['x']),
                                                self.to_int(a['x']), self.to_int(b['z']))

                element_k = self.get_cross_product_for_k(self.to_int(a['x']), self.to_int(b['y']),
                                                self.to_int(a['y']), self.to_int(b['x']))

            return (element_i, element_j, element_k)
        except:
            raise TensorException(
                "The cross product is unable to be generated")

    def get_cross_product_for_i(self, ay, bz, az, by):
        """Calculates the cross product for j
        Args:
        ay (variable)
        bz (variable)
        az (variable)
        by (variable)
        Returns:
        result (variable) The result of calculation"""

        return (ay * bz) - (az * by)

    def get_cross_product_for_j(self, az, bx, ax, bz):
        """Calculates the cross product for j
       Args:
       az (variable)
       bx (variable)
       ax (variable)
       bz (variable)
       Returns:
       result (variable) The result of calculation"""
        return (az * bx) - (ax * bz)

    def get_cross_product_for_k(self, ax, by, ay, bx):
        """Calculates the cross product for k
        Args:
        ax (variable)
        by (variable)
        ay (variable)
        bx (variable)
        Returns:
        result (variable) The result of calculation"""

        return (ax * by) - (ay * bx)

    def subtract_tensor(self):
        """Subtracts the result of tensors of equal size"""

        return self.add_or_subtract_tensor(-1)

    def multply_tensor(self):
        """Given a tensor multiples two tensors to get results
            Returns:
            tensor (list) The result of tensor
            Raises:
            TensorException"""
        try:
            tensor_one_size = self.find_tensor_size(self.tensor_one)
            tensor_two_size = self.find_tensor_size(self.tensor_two)

            # Rule 1 for multiplying. Chec if:
            if tensor_one_size[1] != tensor_two_size[0]:  # cannot proceed
                raise TensorException("Cannot Multiply Tensor as both are not of equal size")

            multiply_tensor = self.create_blank_tensor(tensor_one_size[0], tensor_two_size[1])

            for i in range(tensor_one_size[0]):
                for j in range(tensor_two_size[1]):
                    for k in range(tensor_one_size[1]):
                        multiply_tensor[i][j] += self.tensor_one[i][k] * self.tensor_two[k][j]

            return multiply_tensor
        except:
            raise TensorException(
                "The tensor is unable to be multiplied")
            

    def find_tensor_size(self, tensor):
        """Given a tensor (Matrix) gets the size of the tensor
            Args:
            tensor (list)  The values of the tensor
            Returns:
            result (tuple) The size of the tensor
            Raises:
            TensorException"""

        if len(self.tensor_one) > 0 and len(self.tensor_one[0]) > 0:
            # return the size of the tensor as a tuple (rows, cols)
            return len(tensor), len(tensor[0])
        else:
            return len(tensor), 0


"""Refactored Static Class to Base Class"""

class BaseTensor():
    
    def __init__():
        print("Initialize Base class")
        super().__init__()
    
        
    def to_int(self,value):
        """Given a value converts to int
           Args:
           value (variable)  The values of the tensor
           Returns:
           result (int) The value converted to its integer equivelent"""
        return int(str(value))


    def divide(self,value_one, value_two):
        """Given a value converts to int
           Args:
           value_one (variable)  A given value
           value_two (variable)  A given value
           Returns:
           result (int) The value converted to its integer equivelent"""

        if value_one != 0 and value_two != 0:
            return value_one / value_two


    def pretty_print_tensor(self,tensor):
        """Pretty prints the tensor to screen
               Args:
               tensor (list)  The values of the tensor
               Returns:
               None"""
        for i in range(len(tensor)):
            row_string = ("[")
            for j in range(len(tensor[i])):
                row_string += ("{0:>6},".format(tensor[i][j]))
            print(row_string[:-1], "]")


    def is_two_by_two_tensor(self,tensor):
        """Validates the tensor is of size two by two
           Args:
           tensor (list)  The values of the tensor
           Returns:
           result (bool) The result of whether the tensor is a two by two"""

        return len(tensor) == 2 and len(tensor[0]) == 2


    def is_three_by_one_tensor(self,tensor):
        """Validates the tensor is of size three
        Args:
        tensor (list)  The values of the tensor
        Returns:
        result (bool) The result of whether the tensor is of length three"""

        return len(tensor) == 3


    def tensor_to_variables(self,tensor):
        """Converts a two by two tensor into its component parts or tensors of length three and split into component parts
           Args:
           tensor (list)  The values of the tensor
           Returns:
           result (dictionary) The result of whether the tensor being split out
           Raises:
           TensorException"""

        if self.is_two_by_two_tensor(tensor):
            return {'a': tensor[0][0], 'b': tensor[0][1], 'c': tensor[1][0], 'd': tensor[1][1]}
        elif len(tensor) == 3:
            return {'x': tensor[0], 'y': tensor[1], 'z': tensor[2]}
        else:
            raise TensorException("The tensor is unable to converted to associated variables for use")

    def create_blank_tensor(self,rows, cols):
        """Given a blank tensor of defined rows and columns or rows
            Args:
            rows (variable)  The number of rows to be generated
            columns (variable) The number of columns to be generated
            Returns: (list)   A list containing the results of the inverse operation.
            Raises:
            TensorException"""

        if rows != 0 and cols != 0:
            blank_tensor = [0] * rows  # create <rows> of empty rows
            for i in range(rows):  # for each row
                blank_tensor[i] = [0] * cols  # create <cols> of empty cols
            return blank_tensor
        elif rows != 0:
            return [0] * rows
        else:
            raise TensorException("a blank tensor is unable to be generated")


class TensorTests(unittest.TestCase):

    def test_review_matrix_one_week_three_minus_review_matrix_two_week_three__four_by_four(self):
        actual_matrix = Tensor(review_matrix_one_week_three, review_matrix_two_week_three).subtract_tensor()
        self.assertEqual(actual_matrix, review_add_matrix_result)

    def test_review_matrix_one_week_three_minus_review_matrix_two_week_three__four_by_four(self):
        actual_matrix = Tensor(review_matrix_one_week_three, review_matrix_two_week_three).add_or_subtract_tensor(-1)
        self.assertEqual(actual_matrix, review_subtract_matrix_result)

    def test_review_matrix_one_week_three_multiply_review_matrix_two_week_three_four_by_four(self):
        actual_matrix = Tensor(review_matrix_one_week_three, review_matrix_two_week_three).multply_tensor()
        self.assertEqual(actual_matrix, review_multiply_matrix_result)

    def test_review_matrix_one_week_three_multiply_review_vector_week_three_four_by_one(self):
        actual_matrix = Tensor(review_matrix_one_week_three, review_vector_week_three).multply_tensor()
        self.assertEqual(actual_matrix, review_multiply_by_vector_result)

    def test_determinent_week_four(self):
        actual_determinent = Tensor().get_determinent(week_four_determinent_matrix)
        self.assertEqual(actual_determinent, -3)

    def test_inverse_matrix_week_four(self):
        actual_inverse_matrix = Tensor().get_inverse_tensor(week_four_determinent_matrix)
        self.assertEqual(actual_inverse_matrix, [[-0.6666666666666666, 2.3333333333333335], [1.0, -3.0]])

    def test_cross_product_week_four(self):
        actual_cross_product = Tensor().get_cross_product(week_four_cross_product_tensor_one,
                                                          week_four_cross_product_tensor_two)
        self.assertEqual(actual_cross_product, (-45, -27, 13))

    def test_review_matrix_one_week_four_multiply_review_matrix_two_four_four_by_four(self):
        actual_matrix = Tensor(week_four_matrix_one, week_four_matrix_two).multply_tensor()
        self.assertEqual(actual_matrix, [[147, 66, 162, 59], [68, 32, 57, 15], [249, 114, 324, 139], [86, 42, 137, 69]])

    def test_review_matrix_one_week_four_multiply_review_vector_week_three_four_by_one(self):
        actual_matrix = Tensor(week_four_matrix_one, week_four_vector).multply_tensor()
        self.assertEqual(actual_matrix, [[115], [52], [209], [82]])


if __name__ == '__main__':
    unittest.main(argv=['first-arg-is-ignored'], exit=False)

print(Tensor().get_determinent(week_four_determinent_matrix))
print(Tensor().get_inverse_tensor(week_four_determinent_matrix))
print(Tensor().get_cross_product(week_four_cross_product_tensor_one, week_four_cross_product_tensor_two))
print(Tensor(week_four_matrix_one, week_four_matrix_two).multply_tensor())
print(Tensor(week_four_matrix_one, week_four_vector).multply_tensor())
```
