```python
review_matrix_one_week_three = [[5,4,3,2],[9,7,6,5],[3,9,12,15],[22,18,16,10]]
review_matrix_two_week_three = [[18,56,13,12],[7,6,5,4],[33,91,25,44],[21,19,2,1]]
review_vector_week_three = [[7],[11],[10],[23]]

class matrix:
    
    def __init__(self, matrix_one=[], matrix_two_or_vector=[]):
        self.matrix_one = matrix_one
        self.matrix_two_or_vector = matrix_two_or_vector

    def matrixSize(self, matrix):
        # return the size of the matrix as a tuple (rows, cols)
        return (len(matrix), len(matrix[0]))

    def createBlankMatrix(self, rows, cols):
        # Create an empty matrix of [rows x cols] 
        # Each cell will be set to 0
        blank_matrix = [0] * rows    # create <rows> of empty rows
        for i in range(rows):        # for each row
            blank_matrix[i] = [0] * cols  # create <cols> of empty cols 
        return blank_matrix

    def checkMatrixRowsHaveSameNumberOfCells(self, my_list):
        num = len(my_list[0])
        for i in my_list:
            if(num != len(i)):
                return False
        return True

    def addMatrix(self,operator=1):
        # go no further if Matrices don't match in size 
        matrix1_size = matrixSize(self.matrix_one)
        matrix2_size = matrixSize(self.matrix_two_or_vector)
        if (matrix1_size != matrix2_size):
            return ValueError("Matrices DO NOT MATCH. Cannot perform operation.")
    
        # Create a blank matrix for the answr set all cells to 0
        answer_matrix = createBlankMatrix(matrix1_size[0], matrix1_size[1])
        for i in range(matrix1_size[0]):
            for j in range(matrix1_size[1]):
                answer_matrix[i][j] = self.matrix_one[i][j] + (self.matrix_two_or_vector[i][j]*operator)
              
        return answer_matrix

    def subtractMatrix(self):
        return self.addMatrix(self.matrix_one, self.matrix_two, -1)
   
    def multiplyMatrix(self):
        matrix_one_size = self.matrixSize(self.matrix_one)
        matrix_two_size = self.matrixSize(self.matrix_two)
    
        # Rule 1 for multiplying. Chec if:
        #  number of columns in matrix1 = number of rows in matrix2    
        if (matrix_one_size[1] != matrix_two_size[0]):  # cannot proceed
            return ValueError("Cannot Multiply Matrices")  
    
        # output matrix is rows_m1 x cols_m2
        # create a blank matrix first
        result = createBlankMatrix(matrix_one_size[0], matrix_two_size[1])
       
        for i in range(matrix_one_size[0]):
            for j in range(matrix_two_size[1]):
                for k in range (matrix_one_size[1]):
                    result[i][j] += self.matrix_one[i][k] * self.matrix_two[k][j]
        return result

    def prettyPrintMatrix(self, matrix):
        for i in range(len(matrix)):
            row_string = ("[")
            for j in range (len(matrix[i])):
                row_string +=("{0:>6},".format(matrix[i][j]))
            print(row_string[:-1],"]") # Thanks Niall for remove last comma without needing position check
            
    def testMatrixFunctions(self, operator, matrix1, matrix2, test_description):
        print("\n=== Test: ", test_description, " (", operator, ")")
        print("matrix 1 Dimensions:", str(matrixSize(matrix1)))
        # print(matrix1)
        print("matrix 2 Dimensions:", str(matrixSize(matrix2)))
        # print(matrix2)
        if(operator=="ADD"):
            result = addMatrix(matrix1, matrix2)
        elif(operator=="SUBTRACT"):
            result = subtractMatrix(matrix1, matrix2)
        else:
            result = multiplyMatrix(matrix1, matrix2)
        if(type(result)==str):
            print(result)
        else:
            print("Result Dimensions: ", str(matrixSize(result)))
            prettyPrintMatrix(result)
```
